```python
# Create your first MLP in Keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
# load pima indians dataset
dataset = numpy.loadtxt(r"C:\Users\khata\Downloads\diabetes.csv", delimiter="," ,skiprows=1)
# split into input (X ie dependent variables) and output (Y ie independent variables)
X = dataset[:,0:8]   #0-8 columns are dependent variables  with the 8th column is not included
Y = dataset[:,8]     #8 column is independent variable
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer ='uniform', activation='relu'))
model.add(Dense(8, kernel_initializer ='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer ='uniform', activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=20)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# calculate predictions
predictions = model.predict(X)    # predicting Y only using X
print(predictions)
```

```python
# predictions
labeled_predict = [int(numpy.round(x, 0)) for x in predictions]
print(labeled_predict)
```

```python
print("Rounded type: ", type(labeled_predict)) # rounded is a 'list' class
print("Shape of rounded: ", len(labeled_predict))
print("Dataset type: ", type(dataset)) # numpy array?
print("Shape of dataset: ", dataset.shape)
```

```python
# Turn labeled from a 'list' class into a numpy array
newlabeled = numpy.array(labeled_predict)
print("Rounded type: ", type(newlabeled))
```

```python
# Create a confusion matrix with the actual values and predicted probabilities
confusion_Matrix = pd.crosstab(Y, newlabeled, rownames=['Actual'], colnames=['Predicted'], margins=True)
```
